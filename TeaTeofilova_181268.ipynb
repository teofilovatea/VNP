```python
import pandas as pd
diabetes_prevalence = pd.read_csv(r'C:\Users\teat\Desktop\python\Diabetes_prevalence_WorldBank.csv', sep=',')
diabetes_prevalence.head()
```

```python
columns_to_keep = ['Country Name'] + ['2021']

diabetes_prevalence = diabetes_prevalence[columns_to_keep]

print(diabetes_prevalence)
```

```python
nan_counts = diabetes_prevalence.isnull().sum()

# Display the count of NaN values for each column
print("Number of NaN values in each column:")
print(nan_counts)
```

```python
diabetes_prevalence = diabetes_prevalence.dropna()

print("DataFrame after dropping rows with NaN values:")
print(diabetes_prevalence)

print("\nNumber of rows and columns after cleaning:")
print(diabetes_prevalence.shape)
```

```python
import matplotlib.pyplot as plt

european_countries = [
    'Albania', 'Andorra', 'Armenia', 'Austria', 'Azerbaijan', 'Belarus', 
    'Belgium', 'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 
    'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Georgia', 
    'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 
    'Kosovo', 'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 
    'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia', 
    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 
    'Serbia', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 
    'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City'
]

diabetes_prevalence_europe = diabetes_prevalence[diabetes_prevalence['Country Name'].isin(european_countries)]

diabetes_prevalence_europe.rename(columns={'Country Name': 'Country'}, inplace=True)

print("Filtered DataFrame for European countries:")
print(diabetes_prevalence_europe)

plt.figure(figsize=(14, 8))
diabetes_prevalence_europe.plot(
    x='Country', 
    kind='bar', 
    figsize=(14, 8), 
    title='Diabetes Prevalence in European Countries in 2021',
    ylabel='Diabetes Prevalence (%)',
    xlabel='Country'
)

plt.show()
```

```python
general_availability = pd.read_csv(r'C:\Users\teat\Desktop\python\Genral availability of diabetes testing by HbA1c at the primary health care level_WHO.csv', sep=',')
general_availability.head()
```

```python
import numpy as np
def convert_testing_availability(value):
    if value == 'No':
        return 0
    elif value == 'Yes':
        return 1
    else:
        return np.nan  # Drop 'Don't know' and 'No response'

general_availability['Testing Availability'] = general_availability['Value'].apply(convert_testing_availability)

general_availability_cleaned = general_availability.dropna(subset=['Testing Availability'])
general_availability_cleaned

num_rows = len(general_availability_cleaned)
print(f"Number of rows: {num_rows}")
```

```python
general_availability_cleaned.rename(columns={'Location': 'Country'}, inplace=True)
general_availability_cleaned.head()
```

```python
diabetes_prevalence_and_general_availability = pd.merge(diabetes_prevalence_europe, general_availability_cleaned, on='Country', how='inner')
diabetes_prevalence_and_general_availability.head()
```

```python
diabetes_prevalence_and_general_availability.drop(columns=['Language'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['FactComments'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['DateModified'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['ParentLocationCode'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['Location type'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['Period type'], inplace=True)
diabetes_prevalence_and_general_availability.drop(columns=['IsLatestYear'], inplace=True)
```

```python

diabetes_prevalence_and_general_availability.drop(columns=['IndicatorCode'], inplace=True)
diabetes_prevalence_and_general_availability.rename(columns={'2021': 'Diabetes prevalence'}, inplace=True)
diabetes_prevalence_and_general_availability.head()
```
